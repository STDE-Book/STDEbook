## 1. Without observations

```python
# Plot p_S(s) = (20 + \frac{10}{r}) (s -15) \exp{\left(-(20 + \frac{10}{r}) (s-15)  \right)}   \quad  s >15
s = np.linspace(0,40,100)
a1=0.5
a2=0.2
s0 =15
r=1
p_s = (a1+a2/r)*(s-s0 )*np.exp(-(a1+a2/r)*(s-s0 ))
p_s[s<15]=0
```

```python
plt.figure
plt.plot(s, p_s)

#plt.xlim([-0.1,30])
plt.ylim([0,0.4])
```

```python
# Add mean and maximum

Smean = s0  + 2/(a1+a2)
p_smean = (a1+a2/r)*(Smean-s0 )*np.exp(-(a1+a2/r)*(Smean-s0 ))

print(Smean)
Smax =  s0  + 1/(a1+a2)
p_smax = (a1+a2/r)*(Smax-s0 )*np.exp(-(a1+a2/r)*(Smax-s0 )) 
print(Smax)

plt.figure
plt.plot(s, p_s)
plt.plot(Smax*np.ones((10,1)), np.linspace(0,p_smax,10), '--r')

plt.plot(Smean*np.ones((10,1)), np.linspace(0,p_smean,10), '--g')
plt.ylim([0,0.4])
plt.xlim([10,30])
```

```python
# Computing mean cost

# 1. s_hat=0
# pdf
s = np.linspace(0,40,1000)
delta_s = s[1]-s[0]

a1=0.5
a2=0.2
s0 =15
r=1
p_s = (a1+a2/r)*(s-s0 )*np.exp(-(a1+a2/r)*(s-s0 ))
p_s[s<15]=0


# 2. s_hat=s_max
#cost
shat=Smax
#C = 20*(s-shat)**2
C = (s-shat)

cc= 0.1
C[C>0] = cc*C[C>0]
C[C<0] = -cc/20*C[C<0]

# s_hat
p_shat =  (a1+a2/r)*(shat-s0 )*np.exp(-(a1+a2/r)*(shat-s0 ))  

print(shat)

# product
Cxp_s = p_s*C

Area = np.sum(delta_s*Cxp_s)
print (Area)

plt.figure(2)
plt.plot(s, p_s,'b')
plt.plot(shat*np.ones((10,1)), np.linspace(0,p_shat,10), '--b')
plt.plot(s, C,'g')
plt.plot(s, Cxp_s,'r')
plt.fill_between(s, 0, Cxp_s, facecolor='red')
plt.ylim([0,0.4])
plt.xlim([10,30])

# 3. s_hat=smean
#cost
shat=Smean
#C = 20*(s-shat)**2
C = (s-shat)

cc= 0.1
C[C>0] = cc*C[C>0]
C[C<0] = -cc/20*C[C<0]

# s_hat
p_shat =  (a1+a2/r)*(shat-s0 )*np.exp(-(a1+a2/r)*(shat-s0 ))  
print(shat)

# product
Cxp_s = p_s*C
Area = np.sum(delta_s*Cxp_s)
print (Area)

plt.figure(3)
plt.plot(s, p_s,'b')
plt.plot(shat*np.ones((10,1)), np.linspace(0,p_shat,10), '--b')
plt.plot(s, C,'g')
plt.plot(s, Cxp_s,'r')
plt.fill_between(s, 0, Cxp_s, facecolor='red')
plt.ylim([0,0.4])
plt.xlim([10,30])

# 4. s_hat=2*smean
#cost
shat=22
#C = 20*(s-shat)**2
C = (s-shat)
cc= 0.1
C[C>0] = cc*C[C>0]
C[C<0] = -cc/20*C[C<0]

# s_hat
p_shat =  (a1+a2/r)*(shat-s0 )*np.exp(-(a1+a2/r)*(shat-s0 ))  
print(shat)

# product
Cxp_s = p_s*C
Area = np.sum(delta_s*Cxp_s)
print (Area)

plt.figure(4)
plt.plot(s, p_s,'b')
plt.plot(shat*np.ones((10,1)), np.linspace(0,p_shat,10), '--b')
plt.plot(s, C,'g')
plt.plot(s, Cxp_s,'r')
plt.fill_between(s, 0, Cxp_s, facecolor='red')
plt.ylim([0,0.4])
plt.xlim([10,30])

# 5. s_hat=4*smean
#cost
shat=25
#C = 20*(s-shat)**2
C = (s-shat)
cc= 0.1
C[C>0] = cc*C[C>0]
C[C<0] = -cc/20*C[C<0]

# s_hat
p_shat =  (a1+a2/r)*(shat-s0 )*np.exp(-(a1+a2/r)*(shat-s0 ))  
print(shat)

# product
Cxp_s = p_s*C
Area = np.sum(delta_s*Cxp_s)
print (Area)

plt.figure(5)
plt.plot(s, p_s,'b')
plt.plot(shat*np.ones((10,1)), np.linspace(0,p_shat,10), '--b')
plt.plot(s, C,'g')
plt.plot(s, Cxp_s,'r')
plt.fill_between(s, 0, Cxp_s, facecolor='red')
plt.ylim([0,0.4])
plt.xlim([10,30])
```

```python
# Curve of cost vs s_hat
#cost
list_areas =[]
rang_shat= np.arange(15,30,0.01)
for shat in rang_shat:
    C = (s-shat)
    cc= 0.1
    C[C>0] = cc*C[C>0]
    C[C<0] = -cc/20*C[C<0]

    # s_hat
    p_shat = (a1+a2/r)*(shat-s0 )*np.exp(-(a1+a2/r)*(shat-s0 ))  

    # product
    Cxp_s = p_s*C
    Area = np.sum(delta_s*Cxp_s)
    list_areas.append(Area)

list_areas=np.array(list_areas)
pos=np.argmin(list_areas)
s_hat_opt = rang_shat[pos]
MinCost = list_areas[pos]
plt.figure
plt.plot(rang_shat, list_areas,'b')
plt.plot(s_hat_opt*np.ones((10,1)), np.linspace(0,list_areas[pos],10), '--r')
plt.ylim([0,0.4])
print(s_hat_opt)
print(MinCost)
```
